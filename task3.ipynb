```
x = ds.iloc[:, [0, 1, 2, 3]].values
```

Kmeans

```
from sklearn.cluster import KMeans
```

creating cluster within sum of squares

```
lst=[]
j=range(1,11)
for i in j:
    means=KMeans(n_clusters= i)
    means.fit(x)
    lst.append(means.inertia_)
```

```
import matplotlib.pyplot as plt
%matplotlib inline
# plotting the values in a line graph with points
plt.plot(j,lst,'o-')
# plotted graph title
plt.title('Elbow method shows the optimal number of clusters')
# x-axis label
plt.xlabel('Number of clusters')
# y-axis label
plt.ylabel('Cluster sum of squares')
```

Creating k means classifier

```
kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(x)
```

VISUALIZING

```
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 30, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 30, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 30, c = 'green', label = 'Iris-virginica')
#Plotting the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 30, c = 'black', label = 'Centroids')
plt.legend()
```
