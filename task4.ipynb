```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as dataset
```

Importing data set


```
iris=pd.read_csv('Iris.csv')
```

```
iris.describe()
```

VISUALIZATION

```
sns.pairplot(iris.dropna(),hue='Species')
```

Training of the data set

```
inputs = iris[['SepalLengthCm', 'SepalWidthCm',
                             'PetalLengthCm', 'PetalWidthCm']].values
labels = iris['Species'].values
```

```
from sklearn.model_selection import train_test_split

(training_inputs,testing_inputs,training_classes,testing_classes) = train_test_split(inputs,labels, test_size=0.3, random_state=1)
```

```
# Create decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtclassifier = DecisionTreeClassifier()
```

```
# Training classifier on the training set
dtclassifier.fit(training_inputs, training_classes)
# Validating classifier on the testing set using classification accuracy
dtclassifier.score(testing_inputs, testing_classes)
```
